```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

data = pd.read_csv('annual-working-hours-per-worker.csv')

x = [1950, 1960, 1970, 1980, 1990, 2000] #years I want to graph
y = [] 

with open("annual-working-hours-per-worker.csv", "r") as input_file:
    reader = csv.reader(input_file)
    next(reader) #skip the header name
    for entity, code, year, hours in reader:
        if entity == "Argentina" and (int(year) >= 1950 and int(year) <= 2000 and int(year) % 10 == 0):
            y.append(hours)
  

plt.plot(x, y)
plt.show()
for i in range(6):
    print(f"({x[i]}, {y[i]})")
```
